```python
#Merging the datasets
#used the new rent & geography file 8/30/2021

lease = pd.merge(renewal,
                 days_on_market,
                 on='LeaseDealID', 
                 how='outer')

lease = pd.merge(lease,
                 rent_geo,
                 on='LeaseDealID', 
                 how='outer')

lease = pd.merge(lease,
                 LeaseData,
                 on='LeaseDealID', 
                 how='outer')
```

```python
lease_new = lease.copy()
```

```python
lease_new['PropertyID'] = lease['PropertyID'].astype('object', copy=False)
lease_new['PropertyTypeID'] = lease['PropertyTypeID'].astype('object', copy=False)
lease_new['LocationOccupancyID'] = lease['LocationOccupancyID'].astype('object', copy=False)
lease_new['BuildingRatingID'] = lease['BuildingRatingID'].astype('object', copy=False)
lease_new['LeaseDealID'] = lease['LeaseDealID'].astype('object', copy=False)
```

```python
#Function to change the data type from object to numeric

def change_dtype (df, field):
    df[field] = pd.to_numeric(df[field], errors='coerce')
    
    return df

lease_new = change_dtype(lease_new, 'RateActual')
lease_new = change_dtype(lease_new, 'RBA')
lease_new = change_dtype(lease_new, 'TenantImprovementAllowancePerSqFt')
lease_new = change_dtype(lease_new, 'FreeMonths')
```
