```python
import pandas as pd
import datetime
import numpy as np
pd.set_option('display.max_columns', 500)
import warnings
warnings.filterwarnings('ignore')
```

```python
lease = pd.read_csv('lease_clean_oct29.csv', low_memory=False).drop(columns='Unnamed: 0')
```

```python
print(lease.shape)
```

```python
lease[lease['cbsaid']==35614.0	]
```

```python
# Reading the downloaded file from Kaggle- USA States to region
states = pd.read_csv('states.csv')
print(states.shape)
print(states['Division'].value_counts())
states
```

```python
lease = lease[~(lease['estimated_rent'].isnull())]

#Calculate the current estimated rent-- we want to look at the rent as of today for the submarkets
lease['current_estimated_rent'] = lease['estimated_rent'] * 1.02 ** (datetime.datetime.now().year - lease['year_from'])

lease.cbsa_states =  lease.cbsa_states.str.strip()

```

```python
lease['current_estimated_rent'].max()
```

```python
#Checking the unique states
pd.unique(lease['cbsa_state_new'])
```

```python
#Merge lease and states files
lease_state = pd.merge(lease, states[['State Code', 'Region', 'Division']],
                       left_on='cbsa_state_new', right_on = 'State Code', how='left')

lease_state.drop(columns=['State Code'], inplace=True)
lease_state
```

# Cluster by rent

```python
cluster_df = lease_state.groupby(['cbsaid','submarket_name', 'cbsa_state_new', 'cbsa_cities']).\
    agg({'current_estimated_rent': 'mean'}).reset_index()
cluster_df
```

```python
#by building rating id
rent_mean =  lease_state.groupby(['building_rating_id']).\
    agg({'current_estimated_rent': 'mean'}).reset_index()
rent_mean
```

### Building Clusters

```python
#Expensive high quality building
#Expensive low quality building
#Low cost low quality building
#low cost high quality building

Low_quality = [1,2,3]
High_quality = [4,5]
```

```python
high_quality_mean_rent = lease_state[lease_state['building_rating_id'].isin(High_quality)][['building_rating_id',
                                                                                'current_estimated_rent']]
high_quality_mean_rent.describe().T
```

```python
low_quality_mean_rent = lease_state[lease_state['building_rating_id'].isin(Low_quality)][['building_rating_id',
                                                                                'current_estimated_rent']]
low_quality_mean_rent.describe().T
```

```python
high_quality_expensive = cluster_df[cluster_df['current_estimated_rent'] >35]

#Create column cluster and identify >35 as High Quality Expensive
high_quality_expensive['cluster'] = 'High Quality Expensive'
print(high_quality_expensive.shape)
high_quality_expensive.head(1)
```

```python
#Merge high_quality_expensive with cluster_df
cluster_df = pd.merge(high_quality_expensive[['cbsaid', 'submarket_name', 'cluster']], 
                    cluster_df, on =['cbsaid', 'submarket_name'], how='right' )

cluster_df.head(1)
```

```python
# Create low_quality_low_cost

low_quality_low_cost = cluster_df[(cluster_df['cluster'].isnull()) & \
          (cluster_df['current_estimated_rent'] < 19)]

#Create column cluster and identify <22 as Low Quality Low Cost
low_quality_low_cost['cluster'] = 'Low Quality Low Cost'
print(low_quality_low_cost.shape)

#Merge low_quality_low_cost with cluster_df
cluster_df = pd.merge(low_quality_low_cost[['cbsaid',  'submarket_name','cluster']], 
                    cluster_df, on =['cbsaid',  'submarket_name'], how='right' )

cluster_df['cluster_x'].fillna(cluster_df['cluster_y'], inplace=True)
cluster_df.drop(columns = 'cluster_y' ,inplace=True)
cluster_df.rename(columns={'cluster_x': 'cluster'}, inplace=True)
```

```python
# Create low_quality_expensive
low_quality_expensive = cluster_df[(cluster_df['cluster'].isnull()) & \
          (cluster_df['current_estimated_rent'] <= 30)]
low_quality_expensive['cluster'] = 'Low Quality Expensive'
print(low_quality_expensive.shape)
cluster_df = pd.merge(low_quality_expensive[['cbsaid',  'submarket_name', 'cluster']], 
                    cluster_df, on =['cbsaid', 'submarket_name'], how='right' )

cluster_df['cluster_x'].fillna(cluster_df['cluster_y'], inplace=True)
cluster_df.drop(columns = 'cluster_y' ,inplace=True)
cluster_df.rename(columns={'cluster_x': 'cluster'}, inplace=True)
cluster_df
```

```python
#Create high_quality_low_cost

high_quality_low_cost = cluster_df[(cluster_df['cluster'].isnull()) & \
          (cluster_df['current_estimated_rent'] <= 35)]
high_quality_low_cost['cluster'] = 'High Quality Low Cost'
print(high_quality_low_cost.shape)
cluster_df = pd.merge(high_quality_low_cost[['cbsaid', 'submarket_name','cluster']], 
                    cluster_df, on =['cbsaid',   'submarket_name'], how='right' )

cluster_df['cluster_x'].fillna(cluster_df['cluster_y'], inplace=True)
cluster_df.drop(columns = 'cluster_y' ,inplace=True)
cluster_df.rename(columns={'cluster_x': 'cluster'}, inplace=True)
cluster_df
```
